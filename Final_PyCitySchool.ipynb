```python
# Combine the data into a single dataset.  
#Merge school and students
school_data_complete = pd.merge(student_data, school_data, how="left", on=["school_name", "school_name"])
school_data_complete

#school_index=school_data.set_index('school_name')

#School Name
school_name=school_data['school_name'].unique()

#School Type
# school_type=school_data.groupby(['school_name']).nunique() ["type"]
school_type = school_data.set_index(['school_name'])['type']

#Total students
# students_school=school_data.groupby(['school_name']).unique()['size'] 
students_school = school_data_complete['school_name'].value_counts()

#Budget per school
budget_ps=school_data.groupby(['school_name']).mean()['budget']

#budget per student- add on column
budget_per_student=budget_ps / students_school
school_data['budget_per-student']=budget_per_student

```

```python
# Set school as index and groupby school for merge_df to get average scores by school
index = school_data_complete.set_index('school_name')
schools = index.groupby(['school_name'])
schools
```

```python
# Create an overview table that summarizes key metrics about each school, including:
# Average Math Score
avg_m_school=schools["math_score"].mean()
avg_m_school.head()
```

```python
#groupby to get Average math Score by school

avg_grouped_math=pd.DataFrame(schools['math_score'].mean())
avg_grouped_math.head()
```

```python
# Average Reading Score
avg_r_school=schools["reading_score"].mean()
avg_r_school.head()
```

```python
#groupby to get Average reading Score by school

avg_grouped_reading=pd.DataFrame(schools['reading_score'].mean())
avg_grouped_reading.head()
```

```python
passing_matha=student_data[student_data['math_score']>=70]
passing_matha.head()
```

```python
passing_readinga=student_data[student_data['reading_score']>=70]
passing_readinga.head()
```

```python
# # % Passing Math
total_stu=schools['student_name'].count()
grouped_math=passing_matha.groupby('school_name')
passing_m_school= grouped_math['student_name'].count()/total_stu*100
passing_m_school.head()
```

```python
# # % Passing Reading
grouped_reading=passing_readinga.groupby("school_name")
passing_r_school=grouped_reading['student_name'].count()/total_stu*100
passing_r_school.head()
```

```python
# % Overall Passing (The percentage of students that passed math and reading.)
pass_mr= (passing_m_school+passing_r_school) /2

pass_mr
```

## Scores by School Spending

* Create a table that breaks down school performances based on average Spending Ranges (Per Student). Use 4 reasonable bins to group school spending. Include in the table each of the following:
  * Average Math Score
  * Average Reading Score
  * % Passing Math
  * % Passing Reading
  * Overall Passing Rate (Average of the above two)

```python
# Sample bins. 
#eleventh_grade_perschool_reading=eleventh_grade.groupby(['school_name']).mean()['reading_score']

#Create Bins
spending_bins = [0, 584, 629, 644, 675]
#Names for Bins
spending_range = ["<$584", "$585-629", "$630-644", "$645-675"]

# schools_analysis_df['Per Student Budget']=schools_analysis_df['Per Student Budget'].str.replace('$','')
# schools_analysis_df['Per Student Budget']=schools_analysis_df['Per Student Budget'].astype(float)

schools_analysis_df["Spending Ranges (Per Student)"]=pd.cut(schools_analysis_df['Per Student Budget'],
                                                            spending_bins, labels=spending_range)


spend_summary= schools_analysis_df.groupby(["Spending Ranges (Per Student)"])


spend_summary=spend_summary[['Avg Math Score', "% Passing Math","Avg Reading Score", "% Passing Reading", "% Overall Passing"]]

spend_summary.mean()

#holy cow i did it! 
```
